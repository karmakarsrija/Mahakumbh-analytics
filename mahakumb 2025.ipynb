## Tourism & Economic Impact

```python
#How has the total number of domestic vs. international visitors changed over the years?

visitors_by_year=merged_df.groupby("Year")[
    ["Domestic_Visitors","International_Visitors"]
].sum().reset_index()
visitors_by_year
```

```python
plt.figure(figsize=(10,6))
plt.plot(visitors_by_year["Year"],visitors_by_year["Domestic_Visitors"],marker='o',label="Domestic Visitors")
plt.plot(visitors_by_year["Year"],visitors_by_year["International_Visitors"],marker='o',label="International Visitors")
plt.title("Domestic vs International Visitor over the year")
plt.xlabel("Year")
plt.ylabel("Number of visitors")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Create Total Visitors
merged_df["Total_Visitors"] = merged_df["Domestic_Visitors"] + merged_df["International_Visitors"]

# 2️⃣ Group by Year + Location and sum visitors (if needed)
visitors_by_location = (
    merged_df.groupby(["Year", "Location"])["Total_Visitors"]
    .sum()
    .reset_index()
)

# 3️⃣ Plot: one line per Location
plt.figure(figsize=(12, 6))

for location in visitors_by_location["Location"].unique():
    data = visitors_by_location[visitors_by_location["Location"] == location]
    plt.plot(
        data["Year"],
        data["Total_Visitors"],
        marker='o',
        label=location
    )

plt.title("Total Visitors by Location Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Visitors")
plt.legend(title="Location")
plt.grid(True)
plt.show()
```

```python
max_visitors_row = visitors_by_location.loc[visitors_by_location["Total_Visitors"].idxmax()]

print(f"Year: {max_visitors_row['Year']}")
print(f"Location: {max_visitors_row['Location']}")
print(f"Total Visitors: {max_visitors_row['Total_Visitors']:,}")
```

```python
#Which location generated the highest total revenue?

revenue_by_location=merged_df.groupby("Location")[
    ["Total_Revenue_Generated"]
].sum().reset_index()
```

```python
revenue_by_location = revenue_by_location.sort_values(by="Total_Revenue_Generated", ascending=False)
```

```python
revenue_by_location
```

```python
top_location = revenue_by_location.iloc[0]
print(f"\n Location with highest total revenue: {top_location['Location']} with ₹{top_location['Total_Revenue_Generated']:,.0f}")
```

```python
# Is there a correlation between the number of visitors and hotel/transport revenue?
#Calculate correlations and visualize with scatter plots.

import seaborn as sns
merged_df["Total_Visitors"] = merged_df["Domestic_Visitors"] + merged_df["International_Visitors"]
```

```python
corr_columns = ["Total_Visitors", "Hotel_Revenue", "Transport_Revenue"]

# Calculate correlation matrix
correlation = merged_df[corr_columns].corr()
print("Correlation matrix:")
print(correlation)
```

```python
plt.figure(figsize=(12, 5))

# Visitors vs Hotel Revenue
plt.subplot(1, 2, 1)
sns.scatterplot(data=merged_df, x="Total_Visitors", y="Hotel_Revenue")
sns.regplot(data=merged_df, x="Total_Visitors", y="Hotel_Revenue", scatter=False, color="red")
plt.title("Total Visitors vs Hotel Revenue")

# Visitors vs Transport Revenue
plt.subplot(1, 2, 2)
sns.scatterplot(data=merged_df, x="Total_Visitors", y="Transport_Revenue")
sns.regplot(data=merged_df, x="Total_Visitors", y="Transport_Revenue", scatter=False, color="red")
plt.title("Total Visitors vs Transport Revenue")

plt.tight_layout()
plt.show()
```

The correlation between Total visitors and Hotel revenue is about -0.49(mooderate negative correlation). If, for example, many visitors stay with relatives, in budget accomodation, or in informal lodging during peak events.
The correlation between Total visitors and Transport revenue is -0.18(very weak negative correlation)-> may be poeple use local/free transport option

```python
## average spending per visitor vs hotel/transport revenue

corr_columns = ["Avg_Spending_Per_Visitor", "Hotel_Revenue", "Transport_Revenue"]
correlation = merged_df[corr_columns].corr()
print(" Correlation matrix for Avg Spending:")
print(correlation)
```

```python
plt.figure(figsize=(12, 5))

# Avg Spending vs Hotel Revenue
plt.subplot(1, 2, 1)
sns.scatterplot(data=merged_df, x="Avg_Spending_Per_Visitor", y="Hotel_Revenue")
sns.regplot(data=merged_df, x="Avg_Spending_Per_Visitor", y="Hotel_Revenue", scatter=False, color="red")
plt.title("Avg Spending per Visitor vs Hotel Revenue")

# Avg Spending vs Transport Revenue
plt.subplot(1, 2, 2)
sns.scatterplot(data=merged_df, x="Avg_Spending_Per_Visitor", y="Transport_Revenue")
sns.regplot(data=merged_df, x="Avg_Spending_Per_Visitor", y="Transport_Revenue", scatter=False, color="red")
plt.title("Avg Spending per Visitor vs Transport Revenue")
plt.tight_layout()
plt.show()

```

```python
# 2️⃣ By Location: Facet plots
g = sns.lmplot(
    data=merged_df,
    x="Avg_Spending_Per_Visitor",
    y="Hotel_Revenue",
    col="Location",
    height=4,
    aspect=1,
    scatter_kws={"s": 40},
    line_kws={"color": "red"}
)
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Avg Spending per Visitor vs Hotel Revenue by Location")

g2 = sns.lmplot(
    data=merged_df,
    x="Avg_Spending_Per_Visitor",
    y="Transport_Revenue",
    col="Location",
    height=4,
    aspect=1,
    scatter_kws={"s": 40},
    line_kws={"color": "red"}
)
g2.fig.subplots_adjust(top=0.9)
g2.fig.suptitle("Avg Spending per Visitor vs Transport Revenue by Location")
```

Avg Spending vs Hotel Revenue:

Weak positive correlation (+0.16) — so spending per visitor has a slight positive relationship with hotel revenue, but not strong.

Could mean that hotel revenue depends on other factors too: local supply, number of days stayed, type of accommodation, informal stays, etc.
Avg Spending vs Transport Revenue:

Very weak positive (+0.07) → basically no relationship.

This makes sense if transport spending is not fully linked to how much people spend overall — maybe local transport is fixed-price, subsidized, or many people walk or use free services.

Across all locations, there’s no strong link between how much people spend on average and how much total transport revenue is generated.

Haridwar: Visitor Hotspot but Low Revenue Impact
Highest visitors: Haridwar in 1986

But highest revenue: Prayagraj — so the place with the most visitors is not the top revenue generator.

Insight: Large crowds don’t guarantee high revenue — visitors may be budget-conscious, stay with relatives, use free facilities, or bring their own food.

Follow-up: Could Haridwar increase its revenue by offering better paid services or premium experiences? Targeted hospitality, spiritual tourism packages, or upgraded lodging might help.

✅ 2️⃣ Weak/Negative Visitor–Revenue Correlations
Total Visitors vs Hotel Revenue: Moderate negative (-0.49)
→ More visitors might mean budget travelers or overcrowding → lower per capita hotel spend.

Total Visitors vs Transport Revenue: Very weak negative (-0.18)
→ Suggests many use free/local transport or don’t generate big transport revenue.

Insight: Simply increasing footfall doesn’t guarantee revenue growth in accommodation & transport.

Action: Focus on increasing per visitor spend — e.g., better amenities, transport upgrades, premium transport/tours.

✅ 3️⃣ Average Spending Weakly Linked
Avg Spending vs Hotel Revenue: Weak positive (+0.16)

Avg Spending vs Transport Revenue: Near zero (+0.07)

Insight: People’s spending doesn’t strongly translate into higher hotel or transport earnings — maybe they spend elsewhere (shops, donations, local businesses).

Action: Analyze where visitors actually spend — local markets? souvenirs? food stalls? Then tap those segments for revenue capture or formalization.

✅ 4️⃣ Water Quality Index (WQI): Big Range
Best WQI: 82.6 at Prayagraj (1989) — great!

Worst WQI: 31.7 at Nashik (1980) — poor.

Insight: Large environmental quality variations — could impact visitor satisfaction & safety.

Action: Use this data for policy — e.g., stricter waste management at Nashik, better pollution control → more sustainable events.

✅ 5️⃣ Waste Generation:
Year & place: Nashik in 1980 → highest waste (~4869 tons)

Insight: Massive waste can overwhelm local infrastructure, reduce WQI, increase costs.

Action: Plan scalable waste management — maybe incentivize waste segregation, reuse, eco-friendly disposal.

✅ 6️⃣ Media & Sentiment:
International Media Coverage vs Int’l Visitors: No relationship (-0.04)

Insight: Media buzz doesn’t translate directly to more foreign visitors → maybe due to visas, costs, lack of targeted packages.

Action: Better targeted promotions, or improving international travel ease might work better than just more media.

Positive Sentiment: Varies by year → peaks in 2016 (87.7%), dips in 1992-2013 (~50%).

Insight: Managing public sentiment is key — when sentiment drops, it may affect repeat visits or donations.

✅ 7️⃣ Celebrity Visits & Crime
Highest celebrity visits: Haridwar in 2010.

Crime: Prayagraj highest incidents; correlation with visitors very weak (0.13) → no clear link between crowd size & crime here.

Insight: Crowd management & policing likely keep incidents stable, but hotspots (Prayagraj) may need extra resources.

Action: More targeted safety measures where crime is higher.

📌 Overall Themes
Big picture:

Crowds ≠ guaranteed revenue → quality & spending power matter more than sheer footfall.

Environmental factors (WQI, waste) directly tie into event sustainability & visitor experience.

Sentiment & perception matter — but must be backed by real convenience & offerings.

Celebrities/media create buzz, but alone don’t guarantee international footfall or local revenue.

Public transport & crime are critical for crowd safety but don’t directly drive more revenue.
